```python
import requests

# Define the base URL for the Binance API
BASE_URL = "https://api2.binance.com/api/v3/ticker/price"

# List of 3 symbols for testing (e.g., BTCUSDT, ETHUSDT, BNBUSDT)
symbols = ["BTCUSDT", "ETHUSDT", "BNBUSDT"]

# Function to fetch price for a given symbol
def fetch_price(symbol):
    url = f"{BASE_URL}?symbol={symbol}"
    try:
        response = requests.get(url)
        response.raise_for_status()  # Raise an error for bad status codes
        data = response.json()
        return {
            "symbol": data["symbol"],
            "price": data["price"]
        }
    except requests.exceptions.RequestException as e:
        return {"error": f"Failed to fetch data for {symbol}: {str(e)}"}

# Fetch and print data for each symbol
for symbol in symbols:
    result = fetch_price(symbol)
    print(result)
```

```python
import json
import websocket
import time

# Define the WebSocket URL for combined streams
# For single stream: wss://stream.binance.com:9443/ws/btcusdt@ticker
# For combined: wss://stream.binance.com:9443/stream?streams=btcusdt@ticker/ethusdt@ticker/bnbusdt@ticker
STREAM_URL = "wss://stream.binance.com:9443/stream?streams=btcusdt@ticker/ethusdt@ticker/bnbusdt@ticker"

# Callback function when message is received
def on_message(ws, message):
    data = json.loads(message)
    if 'data' in data:
        ticker = data['data']
        print(f"Symbol: {ticker['s']}, Price: {ticker['c']}, Change: {ticker['P']}%")

# Callback for errors
def on_error(ws, error):
    print(f"Error: {error}")

# Callback when connection closes
def on_close(ws, close_status_code, close_msg):
    print("Connection closed")

# Callback when connection opens
def on_open(ws):
    print("Connection opened")

# Set up WebSocket app
ws_app = websocket.WebSocketApp(
    STREAM_URL,
    on_open=on_open,
    on_message=on_message,
    on_error=on_error,
    on_close=on_close
)

# Run the WebSocket (this will block; in a real app, use threading)
ws_app.run_forever()
```
